```python
#XINLEI HONG ICE2
import pandas as pd
import numpy as np
import string
Eval = pd.read_csv('ICE2_data_eval.csv')
Eval
```

```python
df = pd.DataFrame(Eval)
```

```python
df_concat = pd.DataFrame(columns=["eval", "comment", "category"])

n = df.shape[1]

for i in range(0, n, 2):
  cat = df.columns[i]
  temp_df = pd.DataFrame()
  temp_df["eval"] = df.iloc[:, i]
  temp_df["comment"] = df.iloc[:, i+1]
  temp_df["category"] = cat
  df_concat = pd.concat([df_concat, temp_df])

df_concat
```

```python
Eval_teaching = df[['teaching', 'teaching.1']]
Eval_teaching = Eval_teaching.assign(category = 'teaching')
Eval_teaching.columns =['eval','comment','category']
Eval_teaching
```

```python
Eval_coursecontent = df[['coursecontent', 'coursecontent.1']]
Eval_coursecontent = Eval_coursecontent.assign(category = 'coursecontent')
Eval_coursecontent.columns =['eval','comment','category']
Eval_coursecontent
```

```python
Eval_examination = df[['examination', 'Examination']]
Eval_examination = Eval_examination.assign(category = 'examination')
Eval_examination.columns =['eval','comment','category']
Eval_examination
```

```python
Eval_labwork = df[['labwork', 'labwork.1']]
Eval_labwork = Eval_labwork.assign(category = 'labwork')
Eval_labwork.columns =['eval','comment','category']
Eval_labwork
```

```python
Eval_library_facilities = df.iloc[:,8:10]
Eval_library_facilities = Eval_library_facilities.assign(category = 'library_facilities')
Eval_library_facilities.columns =['eval','comment','category']
Eval_library_facilities
```

```python
Eval_extracurricular = df.iloc[:,10:12]
Eval_extracurricular = Eval_extracurricular.assign(category = 'extracurricular')
Eval_extracurricular.columns =['eval','comment','category']
Eval_extracurricular
```

```python
evalClean = pd.concat([Eval_teaching, Eval_coursecontent,Eval_examination,Eval_labwork,Eval_library_facilities,Eval_extracurricular],ignore_index = True)
evalClean
```

```python
#make text lowercase, 
#remove formats (e.g., html labels), 
#remove punctuation, and 
#remove words containing numbers.
```

```python
import re
```

```python
def clean_text(text):
    #Make text lowercase, remove punctuations, and remove numbers
    text = text.lower()
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('[0-9]+','', text)
    return text
```

```python
evalClean['comment'] = evalClean['comment'].map(clean_text)
evalClean
```

```python
# We are going to create a document-term matrix using CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')

commentCV = cv.fit_transform(evalClean['comment'])
commentCV_dtm = pd.DataFrame(commentCV.toarray(), columns = cv.get_feature_names())

commentCV_dtm.index = evalClean['comment'].index
commentCV_dtm
```

```python
evalClean.groupby(['eval','category']).count().unstack()
```

```python
totalCT = commentCV_dtm.sum()
commentCV_dtm[totalCT.sort_values(ascending = False).index[:20]].sum()
```

```python
from gensim import matutils, models
import scipy.sparse
```

```python
tdm = commentCV_dtm.transpose()
sparse_counts = scipy.sparse.csr_matrix(tdm)
corpus = matutils.Sparse2Corpus(sparse_counts)
id2word = dict((v, p) for p, v in cv.vocabulary_.items())
```

```python
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=3, passes=20)
lda.print_topics()
```

```python
from sklearn.model_selection import train_test_split

Xs_docs = evalClean['comment']
Ys_evals = evalClean['eval']
xs_training, xs_test, y_training, y_test = train_test_split(Xs_docs, Ys_evals, test_size = 0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

# Prepare the training features
cv = CountVectorizer(stop_words='english')
features = cv.fit_transform(xs_training)

# Train a multinomial Naive Bayes Model
model = MultinomialNB()
model.fit(features, y_training)

# Prepare the testing xs.
# Here we are using transform and fit_transform to standardize the data.
# Read about the differences here: https://towardsdatascience.com/what-and-why-behind-fit-transform-vs-transform-in-scikit-learn-78f915cf96fe
feature_test = cv.transform(xs_test)

# Print the model accuracy
print(model.score(feature_test,y_test))
```

```python
from sklearn import svm

# Prepare the training features
cv = CountVectorizer(stop_words='english')
features = cv.fit_transform(xs_training)

# Train SVM classifier
model = svm.SVC()
model.fit(features, y_training)

# Prepare the testing xs
feature_test = cv.transform(xs_test)

# Print the model accuracy
print(model.score(feature_test,y_test))
```
